```python
import pandas as pd
import plotly_express as px
```

```python

olympicdata = pd.read_csv("../Projekt/Data/athlete_events.csv")
```

```python
# Filter with Finland
finland = olympicdata.loc[olympicdata['NOC'].str.contains("FIN", case=False)] 
finland.head(1)
```

```python
# Histogram över ålder

fig = px.histogram(finland, x="Age", color_discrete_sequence = ['blue'], title="Age distribution")
fig.show()
```

```python
# Top sports
wins_by_sport = finland.groupby("Sport").count()
wins_by_sport = wins_by_sport.sort_values("Medal", ascending=False)
top5_sport = wins_by_sport.head(5).reset_index()

fig = px.histogram(top5_sport, x="Sport", y="Medal", title="Top 5 sports")
fig.show()
```

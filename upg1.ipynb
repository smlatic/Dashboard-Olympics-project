```python
import pandas as pd
from Load_data import Load_data
import plotly.express as px
import hashlib

olympic_data = Load_data.load()
olympic_data.head()
```

#### anonymisering av namn

```python

# Get data for finland only
olympic_data_finland = olympic_data.loc[olympic_data['NOC'] == 'FIN']

# Use labmda function to iterate over and hash all names using sha256
olympic_data_finland['Name'] = olympic_data_finland['Name'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())
olympic_data_finland = olympic_data_finland.rename(columns={"Name": "Name (hashed)"})
olympic_data_finland.head()
```

#### De sporter landet f√•tt flest medaljer i

```python
# Get all rows where a player managed to score a medal
olympic_data_finland_with_medals = olympic_data_finland[olympic_data_finland['Medal'].notna()]

# Filter the rows into groups of sports, than sum the "Medals" column for each given sport
medals_across_sports = olympic_data_finland_with_medals.groupby('Sport')['Medal'].value_counts()

# group each sport and sum the medal types (e.g bronze, silver & gold) for each given sport 
medals_per_sport = medals_across_sports.groupby('Sport').sum()

# Sort values, so we get the most medals at the top
medals_per_sport = medals_per_sport.sort_values(ascending=False)

# Take the top 10 sports with most medals scored
medals_per_sport_top10 = medals_per_sport[:10]

fig = px.bar(
    x=medals_per_sport_top10.index,
    y=medals_per_sport_top10,
    title="Medal distribution across the various sports",
    labels={
        'x': '',
        'y': 'Medals'
    }
)

fig.show()
```

#### antal medaljer per OS

```python
# Get all rows where a player managed to score a medal
olympic_data_finland_with_medals = olympic_data_finland[olympic_data_finland['Medal'].notna()]

# Filter the rows into groups of games, than sum the "Medals" column for each given game
medals_across_games = olympic_data_finland_with_medals.groupby('Games')['Medal'].value_counts()

# Group each game and sum the medal types (e.g bronze, silver & gold) for each given game 
medals_per_game = medals_across_games.groupby('Games').sum()

# Sort index, so we can display the games in order of year
medals_per_game.index = medals_per_game.index.sort_values()

fig = px.bar(
    x=medals_per_game.index,
    y=medals_per_game,
    title="Medal distribution across the various olympic games",
    labels={
        'x': '',
        'y': 'Medals'
    }
)

fig.show()
```
